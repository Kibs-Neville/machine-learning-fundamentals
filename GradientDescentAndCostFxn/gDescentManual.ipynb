```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("test_scores.csv")
df
```

```python
plt.figure(figsize = (9,6))
plt.scatter(df.math,df.cs)
plt.xlabel('Score in Maths')
plt.ylabel('Score in CS')
plt.show()
```

```python
x = df['math'].values
y = df['cs'].values
```

```python
x
```

```python
y
```

```python
m = 0
b = 0
epochs = 1000
n = float(len(x))
learning_rate = 0.000001
cost_hist = []

for i in range(epochs):
    y_pred = m*x + b
    cost = (1/n)*sum([val**2 for val in(y_pred-y)])
    cost_hist.append(cost)
    md = (2/n)*sum(x*(y_pred-y))
    bd = (2/n)*sum((y_pred-y))
    m = m-(learning_rate*md)
    b = b-(learning_rate*bd)

plt.plot(range(epochs),cost_hist)
plt.xlabel('Epochs')
plt.ylabel('Cost')
plt.title('Cost Function over Epochs')
plt.show()


```

```python
plt.scatter(x,y, color = 'blue', label = 'Data Points')
plt.plot(x,m*x + b, color = 'red', label = 'Line of best fit')
plt.xlabel('Maths Score')
plt.ylabel('CS Score')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
plt.figure(figsize = (9,6))
plt.grid()
plt.xlabel("Year")
plt.ylabel("Income (US$)")
plt.scatter(df['year'],df['per_capita_income'], color = 'red', marker = '*')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income)
```

## Predicting the price in 2020

```python
x = input("Enter the year you want to predict:")
pred_year = {'year':[x]}
pred_year = pd.DataFrame(pred_year)
reg.predict(pred_year[['year']])
```

```python
reg.coef_
```

```python
reg.intercept_
```

## Our scatterplot with the linear prediction line

```python
plt.figure(figsize = (9,6))
plt.grid()
plt.xlabel("Year")
plt.ylabel("Income (US$)")
plt.scatter(df['year'],df['per_capita_income'], color = 'red', marker = '*')
plt.plot(df.year,reg.predict(df[['year']]), color = 'blue')
plt.show()
```

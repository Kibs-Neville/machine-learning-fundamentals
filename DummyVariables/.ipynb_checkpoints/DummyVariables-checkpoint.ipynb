```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
import joblib
```

```python
df = pd.read_csv("carprices.csv")
```

```python
df
```

```python
df.columns.tolist()
```

## Scatterplot of selling price against age

```python
plt.figure(figsize = (9,6))
plt.scatter(df['Age(yrs)'],df['Sell Price($)'], color = 'blue')
plt.xlabel('Car Age in Years')
plt.ylabel('Car Selling Price')
plt.plot()
```

## Scatterplot of selling price against mileage

```python
plt.figure(figsize = (9,6))
plt.scatter(df['Mileage'],df['Sell Price($)'], color = 'blue')
plt.xlabel('Car Mileage')
plt.ylabel('Car Selling Price')
plt.plot()
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df,dummies], axis = 'columns')
merged
```

```python
final = merged.drop(['Car Model','Audi A5'], axis = 1)
final
```

```python
X = final.drop('Sell Price($)', axis = 1)
```

```python
y = final['Sell Price($)']
```

```python
X
```

```python
y
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[35000,3,1,0]])
```

```python
model.predict([[35000,2,0,0]])
```

```python
accuracy = model.score(X,y)*100
```

```python
print("Model accuracy is:", round(accuracy,2) , "%")
```

```python
joblib.dump(model,'CarModelSellPricePred_joblib')
```
